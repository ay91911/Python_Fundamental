```python
#옆으로 붙이기
df1= pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                 index=[0,1,2,3])

df2=pd.DataFrame({'AA':['A4','A5','A6','A7'],
                  'BB':['B4','B5','B6','B7'],
                  'CC':['C4','C5','C6','C7'],
                  'DD':['D4','D5','D6','D7']},
                 index=[0,1,2,3])

print(df1)
print(df2)
```

```python
#concat디폴트=> 가로로 붙이기,
#세로로 붙이고 싶을땐, axis파라미터 1값 사용
pd.concat([df1,df2],axis=1)
```

```python
#append는 세로로 붙이는 기능이 없음
df1.append(df2,axis=1)
```

```python
import pandas as pd
#가로세로 공통,다른 부분이 각각 있을떄
df1= pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                 index=[0,1,2,3])

df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'CC':['C4','C5','C6','C7'],
                  'DD':['D4','D5','D6','D7']},
                 index=[2,3,5,7])

print(df1)
print(df2)
```

```python
#가로 공통된 부분은 합쳐짐, 세로는 같은 부분 합쳐지지 않음
pd.concat([df1,df2],axis=1)
```

```python
#세로 공통된 부분은 합쳐짐, 가로는 같은 부분 합쳐지지 않음
pd.concat([df1,df2],axis=0)
```

```python
#원하는 부분출력
pd.concat([df1,df2],axis=1).loc[:,["A","B","CC"]]
```

```python
#na값 drop
pd.concat([df1,df2],axis=1).dropna()
```

```python
#join_axes기능 사라짐=>하여 아래와 같이 사용
pd.concat([df1,df2],axis=1).reindex(df2.index)
```

```python
pd.concat([df1,df2],axis=1).reindex(df1.index)
```

```python
#01)DataSet
baseball_player=pd.DataFrame(
    {
        "player_Name":
        ["권혁","이범호","필","박병호","양준혁"],
        "team_No":
        [1993,2001,2001,2010,None]
    }
)
baseball_team=pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
        "team_id":[1993,1999,2001,2010]
    })
```

```python
baseball_player
```

```python
baseball_team
```

```python
pd.merge(baseball_player, baseball_team, how= "inner", left_on="team_No", right_on="team_id")
```

```python
pd.merge(baseball_player, baseball_team, how= "left", left_on="team_No", right_on="team_id")
```

```python
pd.merge(baseball_player, baseball_team, how= "right", left_on="team_No", right_on="team_id")
```

```python
temp= pd.merge(baseball_player, baseball_team, how= "left", left_on="team_No", right_on="team_id")
temp
```

```python
#left outer join-boolean index활용
temp[temp["team_id"].isnull()]
```

```python
temp1=pd.merge(baseball_player, baseball_team, how= "right", left_on="team_No", right_on="team_id")
temp1
```

```python
#right outer join-boolean index활용
temp1[temp1["team_No"].isnull()]
```

```python
#union
pd.merge(baseball_player, baseball_team, how= "outer", left_on="team_No", right_on="team_id")
```

```python
#full outer excluding join
temp2=pd.merge(baseball_player, baseball_team, how= "outer", left_on="team_No", right_on="team_id")
temp2
```

```python
temp2[temp2["team_No" or "team_id"].isnull()]
```

```python
temp2[temp2["team_id"].isnull()|temp2["team_No"].isnull()]
```

```python
baseball_player=pd.DataFrame(
    {
        "player_Name":
        ["권혁","이범호","필","박병호","양준혁"],
        "team_No":
        [1993,2001,2001,2010,None]
    },
    index = ["p_01","p_02","p_03","p_04","p-05"]
)
baseball_player.index.name="player_id"
#player는 기준이 컬럼: team_No, 팀 정보는 기준이 가로 인덱스에 존재!!

baseball_team=pd.DataFrame(
    {
        "team_Name":["한화","두산","기아","넥센"],
        "team_home":["대전","서울","광주","고척"],
    },
    index = [1993,1999,2001,2010]
)
```

```python
baseball_player
```

```python
baseball_team
```

```python
#inner join
pd.merge(baseball_player,baseball_team, how="inner",left_on="team_No", right_index=True)
```

```python
pd.merge(baseball_player,baseball_team, how="left", left_on="team_No", right_index=True)
```

```python
pd.merge(baseball_player,baseball_team, how="right", left_on="team_No", right_index=True)
```

```python
pd.merge(baseball_player,baseball_team, how="outer", left_on="team_No", right_index=True)
```

```python
temp=pd.merge(baseball_player,baseball_team, how="outer", left_on="team_No", right_index=True)
temp[temp.index.isnull()|temp["team_No"].isnull()]
```

```python
#left out join
temp=pd.merge(baseball_player,baseball_team, how="left", left_on="team_No", right_index=True)
temp
```

```python
temp[temp["team_No"].isnull()]
```
