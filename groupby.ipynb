Q) item 값이 call인 데이터에서 통신사별로(network) duration의 값의 총 합을 보고자 함.

```python
data[data["item"]=="call"].groupby("network")["duration"].sum()
```

```python
data.groupby(["month","item"])["date"].count()
```

```python
#기준: month-item
#보고자 하는 값 : duration, network_type, date
#duration->합, network_type->건수, date-> 제일 처음 값
data.groupby(["month","item"]).agg({"duration":"sum","network_type":"count","date":"first"})#->메소드를 모두 문자열로 지정해야함
```

```python
#위의 것과 같이 pivot 테이블로 만들기
pd.pivot_table(data, index=["month","item"],
               aggfunc={"duration":"sum","network_type":"count","date":"first"})#->메소드를 모두 문자열로 지정해야함
```
