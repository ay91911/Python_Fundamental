```python
import pandas as pd
import numpy as np
```

```python

data=pd.read_excel("C:\\Users\\acorn-508\\python\\sampledata\\11_sales-funnel.xlsx")
data.head()
```

```python
data["Status"]=data["Status"].astype("category")#=>알파벳순서..

data["Status"].cat.set_categories(#=>내가 원하는 값 순서대로
["won","pending","presented","declined"],inplace=True)

data.head()
```

```python
data.info()
```

```python
#Name을 가로줄 기준으로 하고, 데이터를 보자
#-->따로 세로줄을 지정하지 않으면 숫자인 컬럼들과 평균값만 들어온다
pd.pivot_table(data, index=["Name"])
```

```python
pd.pivot_table(data, index=["Name","Rep","Manager"])
```

```python
pd.pivot_table(data, index=["Manager","Rep"])
```

```python
pd.pivot_table(data,#데이터
               index=["Manager","Rep"],#가로줄 세팅
               values=["Price"],#세로줄 세팅
               aggfunc=np.sum)# 각 칸을 무엇으로 채울지 결정.. 여기서는 합으로!
```

```python
pd.pivot_table(data,#데이터
               index=["Manager","Rep"],#가로줄 세팅
               values=["Price"],#세로줄 세팅
               aggfunc=[np.mean,len])# 세로줄 중에서도 어떤식으로 표현할건지
```

```python
pd.pivot_table(data,#데이터
               index=["Manager","Rep"],#가로줄 세팅
               values=["Price"],#세로줄 세팅
               aggfunc=np.sum,
               columns=["Product"],# 세로줄에 대한 값을 해당 컬럼으로 또 나눔
               fill_value=0)#NAN값을 0으로 채움
```

```python
pd.pivot_table(data,#데이터
               index=["Manager","Rep"],#가로줄 세팅
               values=["Price","Quantity"],#세로줄 세팅
               aggfunc=np.sum,
               columns=["Product"],# 세로줄에 대한 값을 해당 컬럼으로 또 나눔
               fill_value=0)#NAN값을 0으로 채움
```

```python
pd.pivot_table(data,#데이터
               index=["Manager","Rep","Product"],#가로줄 세팅
               values=["Price","Quantity"],#세로줄 세팅
               aggfunc=np.sum,
               fill_value=0)#NAN값을 0으로 채움
```

```python
pd.pivot_table(data,#데이터
               index=["Manager","Rep","Product"],#가로줄 세팅
               values=["Price","Quantity"],#세로줄 세팅
               aggfunc=[np.sum,np.mean],
               fill_value=0,
               margins=True)
```

```python
#Price에는 aggfunc : np.sum, Quantity에는 aggfunc: len
table=pd.pivot_table(data,#데이터
               index=["Manager","Status"],#가로줄 세팅
               values=["Price","Quantity"],#세로줄 세팅
               aggfunc={"Price":np.sum,"Quantity":len},
               columns=["Product"],
               fill_value=0,
               margins=True)
table
```

```python
table.query('Status==["won","pending"]')
#table.query(Status==["won","pending"])=>문자열로 받아야함.아니면 에러...
```

```python
####crosstab##->pd.crosstab(가로,세로)=>빈도에 초점을 맞추고 있음!
pd.crosstab(data["Name"],data["Price"])
```

```python
pd.crosstab([data["Manager"],data["Rep"]],[data["Price"],data["Quantity"]],
            margins=True)
```

```python
pd.crosstab([data["Manager"],data["Rep"]],[data["Product"]],
            values=data["Price"],
            aggfunc=[len,np.mean]).fillna(0)
```
