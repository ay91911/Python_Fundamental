```python
sampleURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.xml?key=ecf7a82f2dd50dbdb0b54958f35c7d08&itemPerPage=40&openStartDt=2020"
res = urllib.request.urlopen(sampleURL)
soup=BeautifulSoup(res,"html.parser")
```

```python
#soup에 담긴 영화 정보들을 가지고
#1) DF에 영화코드, 영화이름, 영화 영문이름, 1번 감독 이름, 개봉날짜 담기
movieIDDF = pd.DataFrame()
movieIDDF_temp = pd.DataFrame()
movieIDDF_temp = movieIDDF_temp.append({"moviecd":"","movienm":"","movienmen":"","director":"","opendt":""},ignore_index=True)

cd_idx=movieIDDF_temp.columns.get_loc("moviecd")
nm_idx=movieIDDF_temp.columns.get_loc("movienm")
nmen_idx=movieIDDF_temp.columns.get_loc("movienmen")
dt_idx=movieIDDF_temp.columns.get_loc("opendt")
dir_idx=movieIDDF_temp.columns.get_loc("director")

num=len(soup.find_all("movie"))
for i in range(0,num):
    tempInfo=soup.find_all("movie")[i]
    movieIDDF_temp.iat[0,cd_idx] = tempInfo.moviecd.text
    movieIDDF_temp.iat[0,nm_idx] = tempInfo.movienm.text
    movieIDDF_temp.iat[0,nmen_idx] = tempInfo.movienmen.text
    movieIDDF_temp.iat[0,dt_idx] = tempInfo.opendt.text
    if tempInfo.find("directors").find_all("director")!=[]:
        movieIDDF_temp.iat[0,dir_idx] = tempInfo.find("directors").find_all("director")[0].text
    else:
        pass
    
    movieIDDF=movieIDDF.append(movieIDDF_temp, ignore_index=True)
    
    for j in range(0,len(movieIDDF_temp.columns)):
        movieIDDF_temp.iat[0,j] = ""
    
movieIDDF   





```

```python
#2)개봉날짜를 바탕으로 lambda apply활용해서 year/month/dat의 컬럼 추가 생성
for i in range(0,len(movieIDDF)):    
    movieIDDF["year"]=movieIDDF["opendt"].apply(lambda x: str(x)[:4])
    movieIDDF["month"]=movieIDDF["opendt"].apply(lambda x: str(x)[4:6])
    movieIDDF["day"]=movieIDDF["opendt"].apply(lambda x: str(x)[6:])
movieIDDF
```

```python
#3)movicd컬럼을 DF의 reset_index를 활용해서 인덱스로 설정
movieIDDF=movieIDDF.set_index("moviecd")
movieIDDF
```

```python
movieIDDF= movieIDDF.sort_values(by="opendt")
col_list=["movienm","movienmen","director","opendt","year","month","day"]
movieIDDF = movieIDDF.loc[:,col_list]
movieIDDF.head()
```

```python
#4)오늘 한것과 같이 개별 영화에 대한 상세정보(장르, 1번 배우이름, 1번 배우배역, 영화 출연 배우 수)
movieIDDF["type"]=""
movieIDDF["actornm"]=""
movieIDDF["actor_cast"]=""
movieIDDF["actorCnt"]=""


for i in movieIDDF.index:
    
    addURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.xml?key=ecf7a82f2dd50dbdb0b54958f35c7d08&movieCd="+i
    add = urllib.request.urlopen(addURL)
    add_soup=BeautifulSoup(add,"html.parser")
    
    # i번째 영화의 상세 정보...: add_soup
    
    movieIDDF.at[i,"type"]=add_soup.find("typenm").text
    
    if add_soup.find("actors").find_all("actor") !=[]:
        movieIDDF.at[i,"actornm"] = add_soup.find("actors").find_all("peoplenm")[0].text
        movieIDDF.at[i,"actor_cast"] = add_soup.find_all("actor")[0].find("cast").text
        movieIDDF.at[i,"actorCnt"] = len(add_soup.find_all("actor"))
    else:
        pass
    


```
