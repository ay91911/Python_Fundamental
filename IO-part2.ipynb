```python
#모든 정보를 담을 DF
movieIDDF= pd.DataFrame()

#1개 영화 정보들에 대해서 처리할 DF
movieIDDF_temp= pd.DataFrame()

movieIDDF_temp = movieIDDF_temp.append({"movieCd":"","movieNm":"","movieNmEn":"","openDt":"","directorpeopleNM":""}, ignore_index=True)

temp_movcd_idx = movieIDDF_temp.columns.get_loc("movieCd")
temp_movNm_idx = movieIDDF_temp.columns.get_loc("movieNm")
temp_movNmE_idx = movieIDDF_temp.columns.get_loc("movieNmEn")
temp_movOD_idx = movieIDDF_temp.columns.get_loc("openDt")
temp_movdir_idx = movieIDDF_temp.columns.get_loc("directorpeopleNM")

num=len(movieIdPage["movieListResult"]["movieList"])
for i in range(0,num):
    movieIDDF_temp.iat[0,temp_movcd_idx]=movieIdPage["movieListResult"]["movieList"][i]["movieCd"]
    movieIDDF_temp.iat[0,temp_movNm_idx]=movieIdPage["movieListResult"]["movieList"][i]["movieNm"]
    movieIDDF_temp.iat[0,temp_movNmE_idx]=movieIdPage["movieListResult"]["movieList"][i]["movieNmEn"]
    movieIDDF_temp.iat[0,temp_movOD_idx]=movieIdPage["movieListResult"]["movieList"][i]["openDt"]
    if movieIdPage["movieListResult"]["movieList"][i]["directors"]:
        movieIDDF_temp.iat[0,temp_movdir_idx]=movieIdPage["movieListResult"]["movieList"][i]["directors"][0]["peopleNm"]
    else:
        pass

    movieIDDF=movieIDDF.append(movieIDDF_temp, ignore_index=True)

    #1. movieIDDF_temp에 i번쨰 영화정보들을 담기..(코드,이름,영문,개봉일,감독,,)
    #2. movieIDDF_temp에 i번쨰 영화에 대한 정보를 movieIDDF에 추가-->append
    #3. movieIDDF_temp에 담겨져 있는 정보들을 초기화 하는 부분
    
    for j in range(0, len(movieIDDF_temp.columns)):
        movieIDDF_temp.iat[0,j]=""
    
movieIDDF
```

```python
movieIDDF["year"]=movieIDDF["openDt"].apply(lambda x: str(x)[:4])
movieIDDF["month"]=movieIDDF["openDt"].apply(lambda x: str(x)[4:6])
movieIDDF["day"]=movieIDDF["openDt"].apply(lambda x: str(x)[6:])
```

```python
movieIDDF=movieIDDF.sort_values(by="openDt", ascending=True)
movieIDDF.head()
```

```python
movieIDDF=movieIDDF.set_index("movieCd")
movieIDDF.head()
```

```python
col_list=["openDt","year","month","day","movieNm","movieNmEn","directorpeopleNM"]
movieIDDF = movieIDDF.loc[:,col_list]
movieIDDF.head(5)
```

```python
movieIDDF["type"]=""
movieIDDF["ActorNum"]=""
movieIDDF["cast"]=""
movieIDDF["cast_pelpleNm"]=""

type_idx = movieIDDF.columns.get_loc("type")
ActorNum_idx = movieIDDF.columns.get_loc("ActorNum")
cast_idx = movieIDDF.columns.get_loc("cast")
cast_pelpleNm_idx = movieIDDF.columns.get_loc("cast_pelpleNm")
    
for i in range(len(movieIDDF.index)):
    movieInfoURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=ecf7a82f2dd50dbdb0b54958f35c7d08&movieCd="+movieIDDF.index[i]
    movieInfo = urllib.request.urlopen(movieInfoURL)   
    movieInfo = json.loads(movieInfo.read().decode("utf8"),encoding="utf8")
    movieIDDF.iat[i,type_idx]=movieInfo["movieInfoResult"]["movieInfo"]["typeNm"]
    movieIDDF.iat[i,ActorNum_idx]=len(movieInfo["movieInfoResult"]["movieInfo"]["actors"])
    if movieInfo["movieInfoResult"]["movieInfo"]["actors"]:
        movieIDDF.iat[i,cast_idx]=movieInfo["movieInfoResult"]["movieInfo"]["actors"][0]["cast"]
        movieIDDF.iat[i,cast_pelpleNm_idx]=movieInfo["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"]
    else:
        pass
    
movieIDDF





```

```python
#1)기존의 DF에 필요한 컬럼 추가: 장르, 1번 배우 이름, 1번 배우 배역 이름, 출연배우 숫자
#DF["필요한 컬럼이름 추가"]=""
movieIDDF["type"]=""
movieIDDF["ActorNum"]=""
movieIDDF["cast"]=""
movieIDDF["cast_pelpleNm"]=""

#2) 내가 가지고 있는 DF의 가로줄이 몇 개인지 확인(len, shape 등..)
#for 숫자로 돌리기, 값으로 돌리기
#3) for i in ~~~~:
    #1)i번째 영화에 대한 상세정보 요청 주소 만들기
    #2)i번째 영화에 대한 상세정보 주소를 http통신으로 요청하고 받아서 JSON으로 변환
    #3)i번째 영화에 대해서 받은 상세정보를 가지고 필요한 정보 추출 -> i번째 영화 각 컬럼에 붙여넣기


for i in movieIDDF.index:
    movieInfoURL="http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieInfo.json?key=ecf7a82f2dd50dbdb0b54958f35c7d08&movieCd="+i
    movieInfo = urllib.request.urlopen(movieInfoURL)   
    movieInfo = json.loads(movieInfo.read().decode("utf8"),encoding="utf8")
    
    movieIDDF.at[i,"type"]=movieInfo["movieInfoResult"]["movieInfo"]["typeNm"]
    movieIDDF.at[i,"ActorNum"]=len(movieInfo["movieInfoResult"]["movieInfo"]["actors"])
    
    if movieInfo["movieInfoResult"]["movieInfo"]["actors"]:
        movieIDDF.at[i,"cast"]=movieInfo["movieInfoResult"]["movieInfo"]["actors"][0]["cast"]
        movieIDDF.at[i,"cast_pelpleNm"]=movieInfo["movieInfoResult"]["movieInfo"]["actors"][0]["peopleNm"]
    else:
        pass
        
```
